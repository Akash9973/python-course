## Join and Merge operation

#### Merge operation going to check the columns 
#### Join operation is  going to check the rows

# MERGE

```python
data1 = pd.DataFrame({'key1':[1,2,3,4,5],
                     'key2':[1,5,7,2,8],
                     'key3':[91,22,73,45,22],})
```

```python
data2 = pd.DataFrame({'key1':[1,2,32,4,25],
                     'key4':[81,56,78,23,85],
                     'key5':[961,232,773,435,222],})
```

```python
data1
```

```python
data2
```

```python
## INNER JOIN
pd.merge(data1,data2) ## This merge operation is going to find the intersection of both the dataset 
```

```python
## LEFT JOIN  (every element will be join in first table)
pd.merge(data1,data2, how = 'left')
```

```python
## RIGHT JOIN  (every element will be join in second table)
pd.merge(data1,data2, how = 'right')
```

```python
## OUTER JOIN  (every element will be join in both the table)
pd.merge(data1,data2, how = 'outer')
```

```python
## Cross merge (here it will going to combine with both the table)
pd.merge(data1,data2,how='cross')
```
