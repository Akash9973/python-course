# Matplotlib

Matplotlib is a pyhton library which is used in data visualization,graphical plotting etc


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)  ## this will going to plot the graph between x and y 
```

```python
plt.scatter(x,y)
```

```python
## for labeling along x and y axis also for giving the title of the graph
plt.scatter(x,y ,c="red",alpha=0.7, marker='X')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph between x and y")
```

```python
## for printing the grids in the graph
plt.scatter(x,y ,c="red",alpha=0.7, marker='X')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph between x and y")
plt.grid()
```

```python
## for adjusting the size of the graph
plt.figure(figsize=(4,2))
plt.scatter(x,y ,c="red",alpha=0.7, marker='X')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph between x and y")
plt.grid()
```

## for printing bar graphs

```python

x2=['a','b','c','d','e']
y2=np.random.rand(5)
```

```python
plt.bar(x2,y2 , color="pink")
plt.xlabel("company's name")
plt.ylabel("revenue in crores")
```

```python
## for interchaning the axis
plt.barh(x2,y2 , color="pink")
plt.xlabel("company's name")
plt.ylabel("revenue in crores")
```

```python
x3=np.random.rand(50)
y3=np.random.rand(50)
z3=np.random.rand(50)
fig = plt.figure()
ax=fig.add_subplot(projection ="3d")
ax.scatter(x3,y3,z3 ,color='red')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.zlabel("z-axis")
plt.show()
```
